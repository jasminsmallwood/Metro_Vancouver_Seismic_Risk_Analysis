```python
#shows all unique values (and their count) for each column. This is to see which fields are similar and can be combined

def vc(df, *columns):
    """Value counts - shows ALL rows for each column"""
    results = {}
    
    for col in columns:
        vc = df[col].value_counts().sort("count", descending=True)
        results[col] = vc
        
        print(f"\n{'='*60}\n {col} ({len(vc)} unique values)\n{'='*60}")
        
        # Show all rows for this column
        with pl.Config(tbl_rows=-1):
            print(vc)
    
    return results
```

```python
hazus_mapping = {
    # Residential - RES1 (Single-family)
    'RES1': ['house', 'detached', 'SINGLE FAMILY', 'semidetached_house',
             'bungalow', 'cabin', 'houseboat', 'residential'],

    # Residential - RES2 (Mobile Home)
    'RES2': ['static_caravan', 'demountable'],

    # Residential - RES3A-F (Multi-family)
    'RES3': ['apartments', 'terrace', 'townhouse', 'MULTI FAMILY'],

    # Residential - RES4 (Temporary Lodging)
    'RES4': ['hotel', 'motel', 'tent', 'ger'],

    # Residential - RES5 (Institutional Dormitory)
    'RES5': ['dormitory'],

    # Residential - RES6 (Nursing Home)
    'RES6': ['nursing_home'],

    # Commercial - COM1 (Retail)
    'COM1': ['retail', 'mall', 'kiosk', 'restaurant'],

    # Commercial - COM2 (Wholesale)
    'COM2': ['warehouse'],

    # Commercial - COM3 (Personal/Repair Services)
    'COM3': ['service', 'petrol_station', 'toilets'],

    # Commercial - COM4 (Business/Professional)
    'COM4': ['office', 'COMMERCIAL BUILDING', 'commercial', 'mixed_use',
             'MIXED USE', 'business', 'guardhouse', 'gatehouse', 'funeral_hall'],

    # Commercial - COM5 (Banks)
    'COM5': ['bank'],

    # Commercial - COM6 (Hospital)
    'COM6': ['hospital', 'HOSPITAL'],

    # Commercial - COM7 (Medical Office/Clinic)
    'COM7': [],

    # Commercial - COM8 (Transportation)
    'COM8': ['train_station', 'TRANSPORT BUILDING', 'transportation'],

    # Commercial - COM9 (Recreation / Community / Ancillary)
    'COM9': ['RECREATION FACILITY', 'sports_centre', 'sports_hall', 'gym',
             'recreational', 'SKI', 'GOLF', 'stadium', 'grandstand',
             'PARK BUILDING', 'COMMUNITY CENTRE', 'community_hall', 'AMENITY',
             'roof', 'patio', 'gazebo', 'hut', 'arbor', 'picnic_shelter',
             'bandstand', 'treehouse', 'corridor'],

    # Commercial - COM10 (Theater / Culture)
    'COM10': ['theatre', 'cinema', 'museum', 'MUSEUM'],

    # Industrial - IND1 (Heavy)
    'IND1': ['industrial', 'INDUSTRIAL BUILDING', 'factory', 'RECYCLING BUILDING'],

    # Industrial - IND2 (Light)
    'IND2': ['hangar'],

    # Industrial - IND3 (Food/Drugs/Chemicals)
    'IND3': [],

    # Industrial - IND4 (Metals/Minerals Processing)
    'IND4': [],

    # Industrial - IND5 (High Technology)
    'IND5': [],

    # Industrial - IND6 (Construction)
    'IND6': ['construction'],

    # Agriculture - AGR1
    'AGR1': ['barn', 'farm_auxiliary', 'farm', 'FARM BUILDING', 'greenhouse',
             'stable', 'STABLES', 'livestock', 'silo', 'riding_hall',
             'allotment_house'],

    # Religion - REL1
    'REL1': ['church', 'CHURCH', 'temple', 'mosque', 'MOSQUE', 'synagogue',
             'cathedral', 'chapel', 'place_of_worship', 'religious'],

    # Government - GOV1 (General Services)
    'GOV1': ['civic', 'government', 'public', 'MUNICIPAL HALL', 'post_office',
             'OPERATIONS CENTRE', 'UTILITY BUILDING', 'ANIMAL SHELTER'],

    # Government - GOV2 (Military / Emergency)
    'GOV2': ['fire_station', 'FIRE', 'POLICE', 'military', 'bunker'],

    # Education - EDU1 (Schools/Libraries)
    'EDU1': ['school', 'SCHOOL', 'kindergarten', 'library', 'LIBRARY'],

    # Education - EDU2 (Colleges/Universities)
    'EDU2': ['university', 'college'],

    # Other/Ancillary structures (nulled in output)
    'ANCILLARY': ['shed', 'SHED', 'outbuilding', 'boathouse',
                  'container', 'conservatory'],

    # Lifelines / non-building (nulled in output)
    'LIFELINE': ['bridge', 'storage_tank', 'pumping_station', 'tower'],

    # Historic/Ruins (nulled in output)
    'HISTORIC': ['ruins', 'historic'],

    # Demolished/Invalid (nulled in output)
    'INVALID': ['demolished', 'no', 'N/A', '0', 'OTHER',
                'yes;roof', 'yes;church', '']
}

# Flatten the mapping for easier lookup
type_to_hazus = {}
for hazus_class, building_types in hazus_mapping.items():
    for building_type in building_types:
        type_to_hazus[building_type.lower()] = hazus_class

# Create the categorization
df = df.with_columns(
    pl.col('type')
      .str.to_lowercase()
      .str.strip_chars()
      .replace(type_to_hazus, default='UNCATEGORIZED')
      .alias('hazus_class')
)

# NULL-out requested categories
df = df.with_columns(
    pl.when(pl.col('hazus_class').is_in(
        ['UNCATEGORIZED', 'INVALID', 'HISTORIC', 'LIFELINE', 'ANCILLARY']
    ))
    .then(pl.lit(None))
    .otherwise(pl.col('hazus_class'))
    .alias('type_to_specific')
)

# View results
print("HAZUS Classification Summary:")
print(
    df.group_by('type_to_specific')
      .agg(pl.len().alias('count'))
      .sort('count', descending=True)
)

# View sample mappings
print("\nSample mappings:")
print(
    df.select(['type', 'type_to_specific'])
      .unique()
      .sort('type_to_specific')
)
```

Need to split up Manual Class Description, will take extra parsing. 
#### What is the Manual Class? 
##### Residential buildings are grouped into manual classes based quality of design and materials, architectural age, and number of storeys. 

Sources: 
1. https://www.bcassessment.ca/Files/Misc/UserGuide/Residential_Inventory_Extract_External_User_Guide.pdf
2. https://www.bcassessment.ca/Files/Misc/UserGuide/Tables/dataadv12.pdf

```python
# Set Polars to show all rows
pl.Config.set_tbl_rows(-1)

#seeing unique values and count
df.group_by('MANUAL_CLASS_DESCRIPTION').agg(pl.len().alias('count')).sort('count', descending=True)
```

#### *Pattern Observed*
#### [Storeys] [Building Type] - [Architectural Age] - [Build Quality]

#### *Parsing Considerations*
##### 1. Split on the last dash in each string, and work backwards
##### 2. The quality descriptor is always the last segment after a dash
##### 3. Arch Age is *usually* the second to last segment, THOUGH:
##### 4. Some entries are simpler (eg. Strata Townhouse, has no age/quality)
##### 5. Therefore I will extract components based on numbers of parts (based on number of dashes)

MANUAL_CLASS_DESCRIPTION split into:
1. MC_storeys - numerical storey values
2. MC_subtype - building subtype (SFD, Duplex, Strata Townhouse, etc.)
3. MC_age - architectural age (After 1990, Before 1930, etc.)
4. MC_quality - construction quality level (Semi-Custom, Standard, Custom, Basic, etc.)


```python
def parse_manual_class(df):
    """Parse MANUAL_CLASS_DESCRIPTION into components"""
    
    # CRITICAL: Replace "Semi-Custom" with a placeholder BEFORE normalizing dashes
    df = df.with_columns([
        pl.col('MANUAL_CLASS_DESCRIPTION')
          .str.replace_all('Semi-Custom', 'SEMICUSTOM_PLACEHOLDER')  # Protect it
          .str.replace_all(r'\s*-\s*', ' - ')  # Now normalize dashes
          .str.replace_all('SEMICUSTOM_PLACEHOLDER', 'Semi-Custom')  # Restore it
          .alias('_normalized')
    ])
    
    # Extract storeys
    df = df.with_columns([
        pl.col('_normalized')
          .str.extract(r'(\d+\.?\d*)\s*STY', 1)
          .cast(pl.Float64)
          .alias('MC_storeys'),
    ])
    
    # Split into parts
    df = df.with_columns([
        pl.col('_normalized').str.split(' - ').alias('_parts'),
    ])
    
    df = df.with_columns([
        pl.col('_parts').list.len().alias('_n_parts'),
    ])
    
    # Age patterns
    age_patterns = ['Before 1930', 'After 1930', 'After 1960', 'After 1990', 'All Ages', 
                    'Bef 1930', 'Aft 1930', 'Aft 1960', 'Aft 1990']
    age_regex = '|'.join(age_patterns)
    
    # Quality patterns
    quality_patterns = [
        'Semi-Custom', 'Basic', 'Standard', 'Custom',
        'Ave Q', 'Fair Q', 'Good Q', 'Low Q', 'Very Good Q', 'Excellent Q',
        'Lower Qlty', 'Ave Qlty', 'Better Qlty',
        'Average Quality', 'Low Quality', 'Above Average Quality',
        'Minimum', 'Light Duty', 'Light Construction', 'Alternate Light Constr'
    ]
    quality_regex = '|'.join(quality_patterns)
    
    df = df.with_columns([
        # Extract age
        pl.col('_parts')
          .list.eval(pl.element().str.extract(f'^({age_regex})$', 1))
          .list.drop_nulls()
          .list.first()
          .alias('MC_age'),
        
        # Extract quality
        pl.col('_parts')
          .list.eval(pl.element().str.extract(f'^({quality_regex})$', 1))
          .list.drop_nulls()
          .list.first()
          .alias('MC_quality')
    ])
    
    # Build subtype - remove quality with OR without preceding dash
    df = df.with_columns([
        pl.col('_normalized')
          .str.replace_all(r'\d+\.?\d*\s*STY\s*', '')
          .str.replace(f' - ({age_regex})', '')
          .str.replace(f' - ({quality_regex})', '')  # With dash
          .str.replace(f' ({quality_regex})$', '')   # Without dash at end
          .str.strip_chars()
          .str.strip_chars(' -')
          .str.strip_chars()
          .alias('MC_subtype')
    ])
    
    df = df.drop(['_normalized', '_parts', '_n_parts'])
    
    return df
```

|---------------------|-----------|---------------|----------|---------|

$^1$ Special designation for superior construction quality with high-code seismic design level.


```python
# overlap summary table to analyze and combine accordingly
from typing import Optional

def create_overlap_summary_table(
    df: pl.DataFrame,
    col1: str,
    col2: str,
    col3: Optional[str] = None,
    col4: Optional[str] = None,
    col5: Optional[str] = None,
):
    """
    Create a visual summary table of row overlap for up to 5 fields.
    """
    cols = [col1, col2, col3, col4, col5]
    cols = [c for c in cols if c is not None]

    group_exprs = [
        pl.col(c).is_not_null().alias(f"Has_{c}")
        for c in cols
    ]

    return (
        df.group_by(group_exprs)
          .len()
          .sort("len", descending=True)
    )
```
