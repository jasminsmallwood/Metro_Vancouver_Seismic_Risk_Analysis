```python
# overlap summary table to analyze and combine accordingly
from typing import Optional

def create_overlap_summary_table(
    df: pl.DataFrame,
    col1: str,
    col2: str,
    col3: Optional[str] = None,
    col4: Optional[str] = None,
    col5: Optional[str] = None,
):
    """
    Create a visual summary table of row overlap for up to 5 fields.
    """
    cols = [col1, col2, col3, col4, col5]
    cols = [c for c in cols if c is not None]

    group_exprs = [
        pl.col(c).is_not_null().alias(f"Has_{c}")
        for c in cols
    ]

    return (
        df.group_by(group_exprs)
          .len()
          .sort("len", descending=True)
    )
```

1. Building_Area_M2 + TOTAL_FINISHED_AREA_SQFT
---

```python
# 1. Convert units: SQFT to M2 (1 sqft = 0.092903 m²), and rename
# Convert and rename in one step
df = df.with_columns([(pl.col('TOTAL_FINISHED_AREA_SQFT') * 0.092903).alias('TOTAL_FINISHED_AREA_M2')]).drop('TOTAL_FINISHED_AREA_SQFT')
```

```python

create_overlap_summary_table(df, 'Building_Area_M2', 'TOTAL_FINISHED_AREA_M2')
```

```python
#Before combining, let's check schema...
print(df.select(['Building_Area_M2', 'TOTAL_FINISHED_AREA_M2']).schema)
```

```python
# 2. Convert schema
df = df.with_columns([pl.col('Building_Area_M2').cast(pl.Float64, strict=False).alias('Building_Area_M2')])
```

```python
#now let's see co-occuring rows side by side
co_occurring = df.filter(
    (pl.col('Building_Area_M2').is_not_null()) & 
    (pl.col('TOTAL_FINISHED_AREA_M2').is_not_null()) &
    (pl.col('Building_Area_M2') != 0) &
    (pl.col('TOTAL_FINISHED_AREA_M2') != 0)
).select(['Building_Area_M2', 'TOTAL_FINISHED_AREA_M2']).head(20)

print(co_occurring)
```

Okay there seems to be some disparity here... I will ground truth the row where Building_Area_M2 = 642.15, since it deviates furthest from TOTAL_FINISHED_AREA_M2 the most. I will find the corresponding address:

```python
address = df.filter(
    pl.col('Building_Area_M2').cast(pl.Float64, strict=False) == 642.15
).select(['Full_Street_Address', 'Building_Area_M2', 'TOTAL_FINISHED_AREA_M2'])

print(address)
```

*found on rew* - Floors: 15 - therefore second column is correct. 
Let's try Building_Area_M2 = 3923.9952:

```python
address = df.filter(
    pl.col('Building_Area_M2').cast(pl.Float64, strict=False) == 3923.9952
).select(['Full_Street_Address', 'Building_Area_M2', 'TOTAL_FINISHED_AREA_M2'])

print(address)
```

This is a homeless shelter, menaing the first column is correct. I think it's safe to just take the highest value from each column for any of the co-occuring rows. 

```python
# 3. COMBINE: create new column Total_Building_Area_M2 | for co-occurring rows, take the maximum of the two values | for single values, just use that value
df = df.with_columns([
    pl.when(
        pl.col('Building_Area_M2').cast(pl.Float64, strict=False).is_not_null() &
        pl.col('TOTAL_FINISHED_AREA_M2').is_not_null()
    )
    .then(
        pl.max_horizontal([
            pl.col('Building_Area_M2').cast(pl.Float64, strict=False),
            pl.col('TOTAL_FINISHED_AREA_M2')
        ])
    )
    .otherwise(
        pl.coalesce([
            pl.col('Building_Area_M2').cast(pl.Float64, strict=False),
            pl.col('TOTAL_FINISHED_AREA_M2')
        ])
    )
    .alias('building_area_M2')
])

# Verify the results
print(df.select([
    'Building_Area_M2', 
    'TOTAL_FINISHED_AREA_M2', 
    'building_area_M2'
]).head(20))
```

```python
# 4. drop the old columns

df = df.drop(['Building_Area_M2', 'TOTAL_FINISHED_AREA_M2'])
```

3. Height_M + Height_Flr + STOREYS + MC_Storeys
---

```python
# 1. change to consistent units (change M to floors, 3.3 M = 1 flr) and consistent schemas
def height_to_storeys(height_col):
    """Convert Height_M to estimated storeys (3.3m per storey)"""
    return (height_col / 3.3).ceil().cast(pl.Int32)

# Use it:
df = df.with_columns([
    height_to_storeys(pl.col('Height_M')).alias('storeys_from_height')
])
```

```python
# check schemas
df.select(['storeys_from_height', 'Height_Flr', 'STOREYS','MC_storeys']).schema
```

#### MC_Storeys has values of 1.5 storeys. I will combine the other fields as integers first, then change the data type to float64 to combine with MC_storeys

```python
# 1. convert schema to integers (no decimals needed)
df = df.with_columns([
    pl.col('Height_Flr').cast(pl.Int32, strict=False),
    pl.col('STOREYS').cast(pl.Int32, strict=False),
])
```

```python
create_overlap_summary_table(df, 'storeys_from_height', 'Height_Flr', 'STOREYS', 'MC_storeys')
```

```python
# 2. Combine the 4, taking highest value 
df = df.with_columns([
    pl.max_horizontal(['storeys_from_height', 'Height_Flr', 'STOREYS',])
      .alias('building_storeys_int')
])
```

```python
# check schemas
df.select(['building_storeys_int', 'MC_storeys']).schema
```

```python
# 1. since there is 1.5 storeys in MC_storeys, I will convert to float64
df = df.with_columns([
    pl.col('building_storeys_int').cast(pl.Float64, strict=False),
])
```

```python
# 2. take the highest of the columns, since the default seems to be 1

df = df.with_columns(
    pl.max_horizontal(['building_storeys_int', 'MC_storeys']).alias('building_storeys')
)
```

```python
# 3. drop old columns 
df = df.drop(['storeys_from_height', 'Height_Flr', 'STOREYS', 'Height_M'])
```

```python
# 3. drop old columns
df = df.drop(['building_storeys_int', 'MC_storeys'])
```
